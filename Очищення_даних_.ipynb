```python
def clean_and_extract(attr):
    attr_str = str(attr).strip()

    # Якщо "хрестик" або формати типу 45x45 → нормалізуємо
    if "хрестик" in attr_str.lower() or re.search(r'\d+x\d+', attr_str):
        numbers = re.findall(r'[\d.,]+', attr_str)
        if len(numbers) >= 2:
            value = f"{numbers[0]}x{numbers[1]} хрестиків"
            return "Розмір", value
        else:
            return "Розмір", attr_str

    # Якщо є см/мм → теж в Розмір
    elif re.search(r'\d+.*(см|мм)', attr_str.lower()):
        m = re.search(r'([\d.,\*xX ]+)\s*(см|мм)', attr_str.lower())
        value = m.group(0) if m else attr_str
        return "Розмір", value

    # Інакше залишаємо як є (не змінюємо атрибут)
    return attr_str, None

# Застосування
df[["name_attr_clean", "value_attr_extracted"]] = df["name_attr_clean"].apply(
    lambda x: pd.Series(clean_and_extract(x))
)

# Перезаписуємо value_attr для знайдених розмірів
mask = df["value_attr_extracted"].notna()
df.loc[mask, "value_attr"] = df.loc[mask, "value_attr_extracted"]

# Прибираємо допоміжну колонку
df = df.drop(columns=["value_attr_extracted"])
```

```python
# 2. Очистка значень value_attr

# Словник для буквальних замін
REPLACEMENTS = {
    "..": ".",
    "*": "x",
    "Вишивкаxрестиком": "Вишивка xрестиком",
    "Для вишивкиxрестиком": "Для вишивки xрестиком",
    "Дляxардангеру": "Для xардангеру",
}

# Покращений словник regex-замін з урахуванням пробілів та варіацій
REGEX_REPLACEMENTS = {
    r"\bметр(ів|и|а)?\b": "м",
    r"\b(грам(ів)?|гр)\b": "г",
    r"\bштук(и)?\b": "шт",
    r"\b(міліметр(ів|и|а)?|мм|mm)\b": "мм",
    r"\b(сантиметр(ів|и|а)?|см|cm)\b": "см",
    r"\s*[xх]\s*": "x",
    r"(\d+)\.(\d+)": r"\1,\2",
    r"x(рестиків|ардангеру)\b": r" x\1",
}

def multi_replace_series(s: pd.Series, repl: dict) -> pd.Series:
    """
    Виконує одночасні буквальні заміни.
    """
    if not repl:
        return s

    keys = sorted(repl, key=len, reverse=True)
    pattern = re.compile("|".join(map(re.escape, keys)))

    return s.str.replace(pattern, lambda m: repl[m.group(0)], regex=True)

def regex_replace_series(s: pd.Series, repl: dict) -> pd.Series:
    """
    Послідовно виконує regex-заміни.
    """
    for pat, rep in repl.items():
        s = s.str.replace(pat, rep, regex=True)
    return s

# Пайплайн очищення
col = "value_attr"
df[col] = (
    df[col]
    .astype("string")
    .fillna("no_data")
    .str.rstrip(".,")
    .pipe(multi_replace_series, REPLACEMENTS)
    .pipe(regex_replace_series, REGEX_REPLACEMENTS)
    .str.strip()
    .replace("no_data", pd.NA)
)
```

```python
def replace_exact(x, pattern, repl):
    if pd.isna(x):
        return x
    return re.sub(rf"\b{pattern}\b", repl, str(x))


# НОРМАЛІЗАЦІЇ
def normalize_crosses(x: str):
    """45x45 → 45x45 хрестиків (але 7x7 см лишає)"""
    if pd.isna(x):
        return x
    x = str(x)

    if re.search(r"(см|мм)", x.lower()):
        return x

    nums = re.findall(r"\d+(?:,\d+)?", x)
    if len(nums) >= 2:
        return f"{nums[0]}x{nums[1]} хрестиків"
    return x


def clean_sizes(s: str):
    if not isinstance(s, str):
        return s

    s = re.sub(r'\s*[xхXХ×*]\s*', 'x', s)
    s = re.sub(r'[Ø∅⌀]\.?', '', s)

    s = re.sub(r'\s*(см|cм|cm)', ' см', s, flags=re.I)
    s = re.sub(r'(\d+),\s*(\d+)', r'\1.\2', s)     # 1, 5 → 1.5
    s = re.sub(r'(см)(\s*см)', r'\1', s)
    s = re.sub(r'см\s*x\s*(\d+(?:\.\d+)?)\s*см', r'x\1 см', s)
    s = re.sub(r'\s+', ' ', s).strip()
    s = re.sub(r'x\s*(рестики|рестиків|рестика|рестик|р)\b', ' xрестиків', s)

    if re.fullmatch(r'\d+(\.\d+)?x\d+(\.\d+)?(x\d+(\.\d+)?)?', s):
        s += " см"
    return s


#  МАПІНГИ
THREADS_MAPPING = {
    "12": "120 ниток на 10 см",
    "14": "140 ниток на 10 см",
    "16": "160 ниток на 10 см",
}


# ПРАВИЛА
specific_rules = {
    "Розмір елемента": [normalize_crosses],
    "Кількість дизайнів": [lambda x: x.replace("5 шт", "5") if pd.notna(x) else x],
    "Розмір": [
        clean_sizes,
        lambda x: replace_exact(x, "9 cv", "9 см"),
        lambda x: replace_exact(x, "Розмір 6x9 см", "6x9 см"),
        lambda x: replace_exact(x, "1,,59x2.22 см", "1.59x2.22 см"),
        lambda x: replace_exact(x, "ігашок", "іграшок"),
        lambda x: replace_exact(x, "ігашки", "іграшок"),
        lambda x: replace_exact(x, "220x189xрестикі", "220x189 xрестиків"),
        lambda x: replace_exact(x, "35x50мм", "35x50 мм"),
        lambda x: replace_exact(x, "31x50,5 см. см", "31x50,5 см"),
        lambda x: replace_exact(x, "127,x17.7 см", "127x17.7 см"),
        lambda x: replace_exact(x, "40x90 т xрестиків", "40x90 xрестиків"),
        lambda x: replace_exact(x, "13 x9 см", "13x9 см"),
        lambda x: replace_exact(x, "44 x28 см", " 44x28 см")
    ],
    "Довжина": [
        lambda x: re.sub(r"\s+", " ", str(x)).strip() if pd.notna(x) else x,
        lambda x: re.sub(r"(\d),\s+(\d)", r"\1,\2", x) if pd.notna(x) else x,
        lambda x: re.sub(r"(\d)\s*([смм]|мм)", r"\1 \2", x) if pd.notna(x) else x
    ],

    "Кількість елементів": [lambda x: str(x).rstrip("шт") if pd.notna(x) else x],
    "Кількість ниток на 10 cm": [
        lambda x: THREADS_MAPPING.get(x, x) if pd.notna(x) else x,
        lambda x: x.replace("в", "на") if pd.notna(x) else x,
        lambda x: x.replace("у", "на") if pd.notna(x) else x,
        lambda x: replace_exact(x, "16 ниток на 1 см", "160 ниток на 10 см"),
        lambda x: replace_exact(x, "15 ниток на 1 см", "150 ниток на 10 см"),
        lambda x: replace_exact(x, "10-11 ниток на 1 см", "100 ниток на 10 см"),
        lambda x: replace_exact(x, "12 на 1 см", "120 ниток на 10 см"),
        lambda x: replace_exact(x, "8 ниток на 1 см", "80 ниток на 10 см"),
        lambda x: replace_exact(x, "13 ниток на 1 см", "130 ниток на 10 см"),
        lambda x: replace_exact(x, "14 ниток на 1 см", "140 ниток на 10 см")
    ],
    "Кількість": [
        lambda x: re.sub(r"\s*(шт|бобінок|наліпок|шти|мотків|кільце)\s*$", "", str(x)) if pd.notna(x) else x,
        lambda x: replace_exact(x, "1 штука", "1"),
        lambda x: x.replace("28 бобінок +", "28") if pd.notna(x) else x
    ],
    "Вага": [
        lambda x: x.replace("9,08г", "9,08 г") if pd.notna(x) else x
    ],
     "Призначення": [
        lambda x: replace_exact(x, "Для вишивкиxрестиком", "Для вишивки xрестиком")
    ],
    "Склад": [
        lambda x: replace_exact(x, "100% льон", "Льон"),
        lambda x: replace_exact(x, "100% бавовна", "Бавовна"),
        lambda x: replace_exact(x, "100% коноплі", "Коноплі"),
        lambda x: replace_exact(x, "100 % льон", "Льон"),
    ],
    "Техніка": [
        lambda x: replace_exact(x, "Вишивкаxрестиком", "Вишивка xрестиком")
    ]
}


# УНІВЕРСАЛЬНИЙ АПЛАЄР
def apply_rules(value, funcs):
    for f in funcs:
        value = f(value)
    return value


for attr, funcs in specific_rules.items():
    mask = df["name_attr_clean"] == attr
    df.loc[mask, "value_attr"] = df.loc[mask, "value_attr"].map(lambda x: apply_rules(x, funcs))
```

```python
# Пара допоміжних перевірок (heuristics)
def is_size_value(s):
    if pd.isna(s):
        return False
    s = str(s).lower()
    # містить 'см' або 'мм' або 'хрест' або шаблон  NxM (33x43, 33 x 43) або '%'?
    return bool(re.search(r'(\d+[.,]?\d*\s*(см|мм|мм\.))|(\d+[xхX]\d+)|хрест', s))

def is_length_value(s):
    if pd.isna(s): return False
    s = str(s).lower()
    return bool(re.search(r'\d+[.,]?\d*\s*(м|см|мм|мм\.)\b', s))

def is_color_value(s):
    if pd.isna(s): return False
    s = str(s)
    # кольори зазвичай літери, коми, дефіси; якщо в рядку є цифри — менш імовірно колір
    return bool(re.search(r'^[^\d]{2,}$', s.strip())) and (',' in s or '-' in s or len(s.split()) <= 4)

def looks_like_article(s):
    if pd.isna(s): return False
    s = str(s)
    # артикул — короткий буквенно-цифровий код
    return bool(re.match(r'^[A-Za-zА-Яа-я0-9\-_/]{3,20}$', s.strip()))

def summary_by_category(df, category=None, attrs=None, top_n=10):
    """
    Якщо category=None — робимо для всіх категорій.
    attrs — список атрибутів, які хочемо перевірити (name_attr_clean). Якщо None — візьмемо всі.
    Повертає DataFrame з: category, attr, n_products_in_category, n_products_with_attr, pct
    """
    # working copy
    d = df.copy()
    if attrs is None:
        attrs = d['name_attr_clean'].dropna().unique().tolist()

    categories = [category] if category else d['goods_category'].dropna().unique().tolist()
    rows = []
    for cat in categories:
        dcat = d[d['goods_category'] == cat]
        n_products = dcat['urls'].nunique()
        # для кожного товару/атрибута — чи є non-null value
        # групуємо: (urls, name_attr_clean) --> any non-null
        present = (dcat.dropna(subset=['value_attr'])
                   .groupby(['urls','name_attr_clean']).size()
                   .reset_index()[['urls','name_attr_clean']])
        # convert to presence set for quick check
        pres_set = set(tuple(x) for x in present.values)
        for a in attrs:
            # count products that have attribute a
            n_with = len({u for (u,aa) in pres_set if aa == a})
            pct = n_with / n_products * 100 if n_products else 0
            rows.append({'goods_category': cat, 'attribute': a,
                         'n_products_in_category': n_products,
                         'n_products_with_attr': n_with,
                         'pct_with_attr': round(pct,2)})
    res = pd.DataFrame(rows)
    res = res.sort_values(['goods_category','pct_with_attr'], ascending=[True, False])
    res = res[res['n_products_with_attr'] > 0]
    return res

# звіт для категорії
report_threads = summary_by_category(df, category="Вишивка нитками")
print(report_threads.to_string(index=False))
```

```python
# Функція для витягування ціни, старої ціни та знижки
def extract_price_info(price_str):
    if pd.isna(price_str):
        return pd.Series([None, None, None])

    s = price_str.replace(" ", "").replace(",", ".")

    # Основне число (нова ціна)
    price_match = re.search(r'(\d+(?:\.\d+)?)', s)
    price_new = float(price_match.group(1)) if price_match else None

    # Відсоток знижки
    discount_match = re.search(r'-(\d+)%', s)
    discount_percent = int(discount_match.group(1)) if discount_match else None

    # Стара ціна (якщо є після %)
    old_price_match = re.search(r'-(?:\d+)%(\d+(?:\.\d+)?)', s)
    price_old = float(old_price_match.group(1)) if old_price_match else None

    # Якщо ціле число, в int
    if price_new is not None and price_new.is_integer():
        price_new = int(price_new)
    if price_old is not None and price_old.is_integer():
        price_old = int(price_old)

    return pd.Series([price_new, price_old, discount_percent])

df[['price_new', 'price_old', 'discount_percent']] = df['price'].apply(extract_price_info)
df[['price_new', 'price_old', 'discount_percent']] = df[['price_new', 'price_old', 'discount_percent']].round(2)

# Валюта
df["currency"] = df["price"].str.extract(r'([^\d\s,.-]+)$')[0]  # все, що після чисел і пробілів
df["currency"] = df["currency"].str.replace(r'[^a-zA-Zа-яА-Я]+', '', regex=True).str.lower()

df[['price','price_new','price_old','discount_percent','currency']].sample(10)
```
