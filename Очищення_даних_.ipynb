```python
def clean_and_extract(attr):
    attr_str = str(attr).strip()

    # –Ø–∫—â–æ "—Ö—Ä–µ—Å—Ç–∏–∫" –∞–±–æ —Ñ–æ—Ä–º–∞—Ç–∏ —Ç–∏–ø—É 45x45 ‚Üí –Ω–æ—Ä–º–∞–ª—ñ–∑—É—î–º–æ
    if "—Ö—Ä–µ—Å—Ç–∏–∫" in attr_str.lower() or re.search(r'\d+x\d+', attr_str):
        numbers = re.findall(r'[\d.,]+', attr_str)
        if len(numbers) >= 2:
            value = f"{numbers[0]}x{numbers[1]} —Ö—Ä–µ—Å—Ç–∏–∫—ñ–≤"
            return "–†–æ–∑–º—ñ—Ä", value
        else:
            return "–†–æ–∑–º—ñ—Ä", attr_str

    # –Ø–∫—â–æ —î —Å–º/–º–º ‚Üí —Ç–µ–∂ –≤ –†–æ–∑–º—ñ—Ä
    elif re.search(r'\d+.*(—Å–º|–º–º)', attr_str.lower()):
        m = re.search(r'([\d.,\*xX ]+)\s*(—Å–º|–º–º)', attr_str.lower())
        value = m.group(0) if m else attr_str
        return "–†–æ–∑–º—ñ—Ä", value

    # –Ü–Ω–∞–∫—à–µ –∑–∞–ª–∏—à–∞—î–º–æ —è–∫ —î (–Ω–µ –∑–º—ñ–Ω—é—î–º–æ –∞—Ç—Ä–∏–±—É—Ç)
    return attr_str, None

# –ó–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è
df[["name_attr_clean", "value_attr_extracted"]] = df["name_attr_clean"].apply(
    lambda x: pd.Series(clean_and_extract(x))
)

# –ü–µ—Ä–µ–∑–∞–ø–∏—Å—É—î–º–æ value_attr –¥–ª—è –∑–Ω–∞–π–¥–µ–Ω–∏—Ö —Ä–æ–∑–º—ñ—Ä—ñ–≤
mask = df["value_attr_extracted"].notna()
df.loc[mask, "value_attr"] = df.loc[mask, "value_attr_extracted"]

# –ü—Ä–∏–±–∏—Ä–∞—î–º–æ –¥–æ–ø–æ–º—ñ–∂–Ω—É –∫–æ–ª–æ–Ω–∫—É
df = df.drop(columns=["value_attr_extracted"])
```

```python
# 2. –û—á–∏—Å—Ç–∫–∞ –∑–Ω–∞—á–µ–Ω—å value_attr

# –°–ª–æ–≤–Ω–∏–∫ –¥–ª—è –±—É–∫–≤–∞–ª—å–Ω–∏—Ö –∑–∞–º—ñ–Ω
REPLACEMENTS = {
    "..": ".",
    "*": "x",
    "–í–∏—à–∏–≤–∫–∞x—Ä–µ—Å—Ç–∏–∫–æ–º": "–í–∏—à–∏–≤–∫–∞ x—Ä–µ—Å—Ç–∏–∫–æ–º",
    "–î–ª—è –≤–∏—à–∏–≤–∫–∏x—Ä–µ—Å—Ç–∏–∫–æ–º": "–î–ª—è –≤–∏—à–∏–≤–∫–∏ x—Ä–µ—Å—Ç–∏–∫–æ–º",
    "–î–ª—èx–∞—Ä–¥–∞–Ω–≥–µ—Ä—É": "–î–ª—è x–∞—Ä–¥–∞–Ω–≥–µ—Ä—É",
}

# –ü–æ–∫—Ä–∞—â–µ–Ω–∏–π —Å–ª–æ–≤–Ω–∏–∫ regex-–∑–∞–º—ñ–Ω –∑ —É—Ä–∞—Ö—É–≤–∞–Ω–Ω—è–º –ø—Ä–æ–±—ñ–ª—ñ–≤ —Ç–∞ –≤–∞—Ä—ñ–∞—Ü—ñ–π
REGEX_REPLACEMENTS = {
    r"\b–º–µ—Ç—Ä(—ñ–≤|–∏|–∞)?\b": "–º",
    r"\b(–≥—Ä–∞–º(—ñ–≤)?|–≥—Ä)\b": "–≥",
    r"\b—à—Ç—É–∫(–∏)?\b": "—à—Ç",
    r"\b(–º—ñ–ª—ñ–º–µ—Ç—Ä(—ñ–≤|–∏|–∞)?|–º–º|mm)\b": "–º–º",
    r"\b(—Å–∞–Ω—Ç–∏–º–µ—Ç—Ä(—ñ–≤|–∏|–∞)?|—Å–º|cm)\b": "—Å–º",
    r"\s*[x—Ö]\s*": "x",
    r"(\d+)\.(\d+)": r"\1,\2",
    r"x(—Ä–µ—Å—Ç–∏–∫—ñ–≤|–∞—Ä–¥–∞–Ω–≥–µ—Ä—É)\b": r" x\1",
}

def multi_replace_series(s: pd.Series, repl: dict) -> pd.Series:
    """
    –í–∏–∫–æ–Ω—É—î –æ–¥–Ω–æ—á–∞—Å–Ω—ñ –±—É–∫–≤–∞–ª—å–Ω—ñ –∑–∞–º—ñ–Ω–∏.
    """
    if not repl:
        return s

    keys = sorted(repl, key=len, reverse=True)
    pattern = re.compile("|".join(map(re.escape, keys)))

    return s.str.replace(pattern, lambda m: repl[m.group(0)], regex=True)

def regex_replace_series(s: pd.Series, repl: dict) -> pd.Series:
    """
    –ü–æ—Å–ª—ñ–¥–æ–≤–Ω–æ –≤–∏–∫–æ–Ω—É—î regex-–∑–∞–º—ñ–Ω–∏.
    """
    for pat, rep in repl.items():
        s = s.str.replace(pat, rep, regex=True)
    return s

# –ü–∞–π–ø–ª–∞–π–Ω –æ—á–∏—â–µ–Ω–Ω—è
col = "value_attr"
df[col] = (
    df[col]
    .astype("string")
    .fillna("no_data")
    .str.rstrip(".,")
    .pipe(multi_replace_series, REPLACEMENTS)
    .pipe(regex_replace_series, REGEX_REPLACEMENTS)
    .str.strip()
    .replace("no_data", pd.NA)
)
```

```python
def replace_exact(x, pattern, repl):
    if pd.isna(x):
        return x
    return re.sub(rf"\b{pattern}\b", repl, str(x))


# –ù–û–†–ú–ê–õ–Ü–ó–ê–¶–Ü–á
def normalize_crosses(x: str):
    """45x45 ‚Üí 45x45 —Ö—Ä–µ—Å—Ç–∏–∫—ñ–≤ (–∞–ª–µ 7x7 —Å–º –ª–∏—à–∞—î)"""
    if pd.isna(x):
        return x
    x = str(x)

    if re.search(r"(—Å–º|–º–º)", x.lower()):
        return x

    nums = re.findall(r"\d+(?:,\d+)?", x)
    if len(nums) >= 2:
        return f"{nums[0]}x{nums[1]} —Ö—Ä–µ—Å—Ç–∏–∫—ñ–≤"
    return x


def clean_sizes(s: str):
    if not isinstance(s, str):
        return s

    s = re.sub(r'\s*[x—ÖX–•√ó*]\s*', 'x', s)
    s = re.sub(r'[√ò‚àÖ‚åÄ]\.?', '', s)

    s = re.sub(r'\s*(—Å–º|c–º|cm)', ' —Å–º', s, flags=re.I)
    s = re.sub(r'(\d+),\s*(\d+)', r'\1.\2', s)     # 1, 5 ‚Üí 1.5
    s = re.sub(r'(—Å–º)(\s*—Å–º)', r'\1', s)
    s = re.sub(r'—Å–º\s*x\s*(\d+(?:\.\d+)?)\s*—Å–º', r'x\1 —Å–º', s)
    s = re.sub(r'\s+', ' ', s).strip()
    s = re.sub(r'x\s*(—Ä–µ—Å—Ç–∏–∫–∏|—Ä–µ—Å—Ç–∏–∫—ñ–≤|—Ä–µ—Å—Ç–∏–∫–∞|—Ä–µ—Å—Ç–∏–∫|—Ä)\b', ' x—Ä–µ—Å—Ç–∏–∫—ñ–≤', s)

    if re.fullmatch(r'\d+(\.\d+)?x\d+(\.\d+)?(x\d+(\.\d+)?)?', s):
        s += " —Å–º"
    return s


#  –ú–ê–ü–Ü–ù–ì–ò
THREADS_MAPPING = {
    "12": "120 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º",
    "14": "140 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º",
    "16": "160 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º",
}


# –ü–†–ê–í–ò–õ–ê
specific_rules = {
    "–†–æ–∑–º—ñ—Ä –µ–ª–µ–º–µ–Ω—Ç–∞": [normalize_crosses],
    "–ö—ñ–ª—å–∫—ñ—Å—Ç—å –¥–∏–∑–∞–π–Ω—ñ–≤": [lambda x: x.replace("5 —à—Ç", "5") if pd.notna(x) else x],
    "–†–æ–∑–º—ñ—Ä": [
        clean_sizes,
        lambda x: replace_exact(x, "9 cv", "9 —Å–º"),
        lambda x: replace_exact(x, "–†–æ–∑–º—ñ—Ä 6x9 —Å–º", "6x9 —Å–º"),
        lambda x: replace_exact(x, "1,,59x2.22 —Å–º", "1.59x2.22 —Å–º"),
        lambda x: replace_exact(x, "—ñ–≥–∞—à–æ–∫", "—ñ–≥—Ä–∞—à–æ–∫"),
        lambda x: replace_exact(x, "—ñ–≥–∞—à–∫–∏", "—ñ–≥—Ä–∞—à–æ–∫"),
        lambda x: replace_exact(x, "220x189x—Ä–µ—Å—Ç–∏–∫—ñ", "220x189 x—Ä–µ—Å—Ç–∏–∫—ñ–≤"),
        lambda x: replace_exact(x, "35x50–º–º", "35x50 –º–º"),
        lambda x: replace_exact(x, "31x50,5 —Å–º. —Å–º", "31x50,5 —Å–º"),
        lambda x: replace_exact(x, "127,x17.7 —Å–º", "127x17.7 —Å–º"),
        lambda x: replace_exact(x, "40x90 —Ç x—Ä–µ—Å—Ç–∏–∫—ñ–≤", "40x90 x—Ä–µ—Å—Ç–∏–∫—ñ–≤"),
        lambda x: replace_exact(x, "13 x9 —Å–º", "13x9 —Å–º"),
        lambda x: replace_exact(x, "44 x28 —Å–º", " 44x28 —Å–º")
    ],
    "–î–æ–≤–∂–∏–Ω–∞": [
        lambda x: re.sub(r"\s+", " ", str(x)).strip() if pd.notna(x) else x,
        lambda x: re.sub(r"(\d),\s+(\d)", r"\1,\2", x) if pd.notna(x) else x,
        lambda x: re.sub(r"(\d)\s*([—Å–º–º]|–º–º)", r"\1 \2", x) if pd.notna(x) else x
    ],

    "–ö—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤": [lambda x: str(x).rstrip("—à—Ç") if pd.notna(x) else x],
    "–ö—ñ–ª—å–∫—ñ—Å—Ç—å –Ω–∏—Ç–æ–∫ –Ω–∞ 10 cm": [
        lambda x: THREADS_MAPPING.get(x, x) if pd.notna(x) else x,
        lambda x: x.replace("–≤", "–Ω–∞") if pd.notna(x) else x,
        lambda x: x.replace("—É", "–Ω–∞") if pd.notna(x) else x,
        lambda x: replace_exact(x, "16 –Ω–∏—Ç–æ–∫ –Ω–∞ 1 —Å–º", "160 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º"),
        lambda x: replace_exact(x, "15 –Ω–∏—Ç–æ–∫ –Ω–∞ 1 —Å–º", "150 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º"),
        lambda x: replace_exact(x, "10-11 –Ω–∏—Ç–æ–∫ –Ω–∞ 1 —Å–º", "100 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º"),
        lambda x: replace_exact(x, "12 –Ω–∞ 1 —Å–º", "120 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º"),
        lambda x: replace_exact(x, "8 –Ω–∏—Ç–æ–∫ –Ω–∞ 1 —Å–º", "80 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º"),
        lambda x: replace_exact(x, "13 –Ω–∏—Ç–æ–∫ –Ω–∞ 1 —Å–º", "130 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º"),
        lambda x: replace_exact(x, "14 –Ω–∏—Ç–æ–∫ –Ω–∞ 1 —Å–º", "140 –Ω–∏—Ç–æ–∫ –Ω–∞ 10 —Å–º")
    ],
    "–ö—ñ–ª—å–∫—ñ—Å—Ç—å": [
        lambda x: re.sub(r"\s*(—à—Ç|–±–æ–±—ñ–Ω–æ–∫|–Ω–∞–ª—ñ–ø–æ–∫|—à—Ç–∏|–º–æ—Ç–∫—ñ–≤|–∫—ñ–ª—å—Ü–µ)\s*$", "", str(x)) if pd.notna(x) else x,
        lambda x: replace_exact(x, "1 —à—Ç—É–∫–∞", "1"),
        lambda x: x.replace("28 –±–æ–±—ñ–Ω–æ–∫ +", "28") if pd.notna(x) else x
    ],
    "–í–∞–≥–∞": [
        lambda x: x.replace("9,08–≥", "9,08 –≥") if pd.notna(x) else x
    ],
     "–ü—Ä–∏–∑–Ω–∞—á–µ–Ω–Ω—è": [
        lambda x: replace_exact(x, "–î–ª—è –≤–∏—à–∏–≤–∫–∏x—Ä–µ—Å—Ç–∏–∫–æ–º", "–î–ª—è –≤–∏—à–∏–≤–∫–∏ x—Ä–µ—Å—Ç–∏–∫–æ–º")
    ],
    "–°–∫–ª–∞–¥": [
        lambda x: replace_exact(x, "100% –ª—å–æ–Ω", "–õ—å–æ–Ω"),
        lambda x: replace_exact(x, "100% –±–∞–≤–æ–≤–Ω–∞", "–ë–∞–≤–æ–≤–Ω–∞"),
        lambda x: replace_exact(x, "100% –∫–æ–Ω–æ–ø–ª—ñ", "–ö–æ–Ω–æ–ø–ª—ñ"),
        lambda x: replace_exact(x, "100 % –ª—å–æ–Ω", "–õ—å–æ–Ω"),
    ],
    "–¢–µ—Ö–Ω—ñ–∫–∞": [
        lambda x: replace_exact(x, "–í–∏—à–∏–≤–∫–∞x—Ä–µ—Å—Ç–∏–∫–æ–º", "–í–∏—à–∏–≤–∫–∞ x—Ä–µ—Å—Ç–∏–∫–æ–º")
    ]
}


# –£–ù–Ü–í–ï–†–°–ê–õ–¨–ù–ò–ô –ê–ü–õ–ê–Ñ–†
def apply_rules(value, funcs):
    for f in funcs:
        value = f(value)
    return value


for attr, funcs in specific_rules.items():
    mask = df["name_attr_clean"] == attr
    df.loc[mask, "value_attr"] = df.loc[mask, "value_attr"].map(lambda x: apply_rules(x, funcs))
```

```python
# –ü–∞—Ä–∞ –¥–æ–ø–æ–º—ñ–∂–Ω–∏—Ö –ø–µ—Ä–µ–≤—ñ—Ä–æ–∫ (heuristics)
def is_size_value(s):
    if pd.isna(s):
        return False
    s = str(s).lower()
    # –º—ñ—Å—Ç–∏—Ç—å '—Å–º' –∞–±–æ '–º–º' –∞–±–æ '—Ö—Ä–µ—Å—Ç' –∞–±–æ —à–∞–±–ª–æ–Ω  NxM (33x43, 33 x 43) –∞–±–æ '%'?
    return bool(re.search(r'(\d+[.,]?\d*\s*(—Å–º|–º–º|–º–º\.))|(\d+[x—ÖX]\d+)|—Ö—Ä–µ—Å—Ç', s))

def is_length_value(s):
    if pd.isna(s): return False
    s = str(s).lower()
    return bool(re.search(r'\d+[.,]?\d*\s*(–º|—Å–º|–º–º|–º–º\.)\b', s))

def is_color_value(s):
    if pd.isna(s): return False
    s = str(s)
    # –∫–æ–ª—å–æ—Ä–∏ –∑–∞–∑–≤–∏—á–∞–π –ª—ñ—Ç–µ—Ä–∏, –∫–æ–º–∏, –¥–µ—Ñ—ñ—Å–∏; —è–∫—â–æ –≤ —Ä—è–¥–∫—É —î —Ü–∏—Ñ—Ä–∏ ‚Äî –º–µ–Ω—à —ñ–º–æ–≤—ñ—Ä–Ω–æ –∫–æ–ª—ñ—Ä
    return bool(re.search(r'^[^\d]{2,}$', s.strip())) and (',' in s or '-' in s or len(s.split()) <= 4)

def looks_like_article(s):
    if pd.isna(s): return False
    s = str(s)
    # –∞—Ä—Ç–∏–∫—É–ª ‚Äî –∫–æ—Ä–æ—Ç–∫–∏–π –±—É–∫–≤–µ–Ω–Ω–æ-—Ü–∏—Ñ—Ä–æ–≤–∏–π –∫–æ–¥
    return bool(re.match(r'^[A-Za-z–ê-–Ø–∞-—è0-9\-_/]{3,20}$', s.strip()))

def summary_by_category(df, category=None, attrs=None, top_n=10):
    """
    –Ø–∫—â–æ category=None ‚Äî —Ä–æ–±–∏–º–æ –¥–ª—è –≤—Å—ñ—Ö –∫–∞—Ç–µ–≥–æ—Ä—ñ–π.
    attrs ‚Äî —Å–ø–∏—Å–æ–∫ –∞—Ç—Ä–∏–±—É—Ç—ñ–≤, —è–∫—ñ —Ö–æ—á–µ–º–æ –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ (name_attr_clean). –Ø–∫—â–æ None ‚Äî –≤—ñ–∑—å–º–µ–º–æ –≤—Å—ñ.
    –ü–æ–≤–µ—Ä—Ç–∞—î DataFrame –∑: category, attr, n_products_in_category, n_products_with_attr, pct
    """
    # working copy
    d = df.copy()
    if attrs is None:
        attrs = d['name_attr_clean'].dropna().unique().tolist()

    categories = [category] if category else d['goods_category'].dropna().unique().tolist()
    rows = []
    for cat in categories:
        dcat = d[d['goods_category'] == cat]
        n_products = dcat['urls'].nunique()
        # –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ —Ç–æ–≤–∞—Ä—É/–∞—Ç—Ä–∏–±—É—Ç–∞ ‚Äî —á–∏ —î non-null value
        # –≥—Ä—É–ø—É—î–º–æ: (urls, name_attr_clean) --> any non-null
        present = (dcat.dropna(subset=['value_attr'])
                   .groupby(['urls','name_attr_clean']).size()
                   .reset_index()[['urls','name_attr_clean']])
        # convert to presence set for quick check
        pres_set = set(tuple(x) for x in present.values)
        for a in attrs:
            # count products that have attribute a
            n_with = len({u for (u,aa) in pres_set if aa == a})
            pct = n_with / n_products * 100 if n_products else 0
            rows.append({'goods_category': cat, 'attribute': a,
                         'n_products_in_category': n_products,
                         'n_products_with_attr': n_with,
                         'pct_with_attr': round(pct,2)})
    res = pd.DataFrame(rows)
    res = res.sort_values(['goods_category','pct_with_attr'], ascending=[True, False])
    res = res[res['n_products_with_attr'] > 0]
    return res

# –∑–≤—ñ—Ç –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó
report_threads = summary_by_category(df, category="–í–∏—à–∏–≤–∫–∞ –Ω–∏—Ç–∫–∞–º–∏")
print(report_threads.to_string(index=False))
```

```python
# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –≤–∏—Ç—è–≥—É–≤–∞–Ω–Ω—è —Ü—ñ–Ω–∏, —Å—Ç–∞—Ä–æ—ó —Ü—ñ–Ω–∏ —Ç–∞ –∑–Ω–∏–∂–∫–∏
def extract_price_info(price_str):
    if pd.isna(price_str):
        return pd.Series([None, None, None])

    s = price_str.replace(" ", "").replace(",", ".")

    # –û—Å–Ω–æ–≤–Ω–µ —á–∏—Å–ª–æ (–Ω–æ–≤–∞ —Ü—ñ–Ω–∞)
    price_match = re.search(r'(\d+(?:\.\d+)?)', s)
    price_new = float(price_match.group(1)) if price_match else None

    # –í—ñ–¥—Å–æ—Ç–æ–∫ –∑–Ω–∏–∂–∫–∏
    discount_match = re.search(r'-(\d+)%', s)
    discount_percent = int(discount_match.group(1)) if discount_match else None

    # –°—Ç–∞—Ä–∞ —Ü—ñ–Ω–∞ (—è–∫—â–æ —î –ø—ñ—Å–ª—è %)
    old_price_match = re.search(r'-(?:\d+)%(\d+(?:\.\d+)?)', s)
    price_old = float(old_price_match.group(1)) if old_price_match else None

    # –Ø–∫—â–æ —Ü—ñ–ª–µ —á–∏—Å–ª–æ, –≤ int
    if price_new is not None and price_new.is_integer():
        price_new = int(price_new)
    if price_old is not None and price_old.is_integer():
        price_old = int(price_old)

    return pd.Series([price_new, price_old, discount_percent])

df[['price_new', 'price_old', 'discount_percent']] = df['price'].apply(extract_price_info)
df[['price_new', 'price_old', 'discount_percent']] = df[['price_new', 'price_old', 'discount_percent']].round(2)

# –í–∞–ª—é—Ç–∞
df["currency"] = df["price"].str.extract(r'([^\d\s,.-]+)$')[0]  # –≤—Å–µ, —â–æ –ø—ñ—Å–ª—è —á–∏—Å–µ–ª —ñ –ø—Ä–æ–±—ñ–ª—ñ–≤
df["currency"] = df["currency"].str.replace(r'[^a-zA-Z–∞-—è–ê-–Ø]+', '', regex=True).str.lower()

df[['price','price_new','price_old','discount_percent','currency']].sample(10)
```

```python
# –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∫–æ–ª–æ–Ω–æ–∫ —É category
df["name_attr_clean"] = df["name_attr_clean"].astype("string")
df["goods_category"] = df["goods_category"].astype("category")
df["goods_status"] = df["goods_status"].map({"–Ñ –≤ –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ": True, "–ù–µ–º–∞—î –≤ –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ": False}).astype("boolean")
df["goods_name"] = df["goods_name"].astype("string")
df["currency"] = df["currency"].astype("string")
```
